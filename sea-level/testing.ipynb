```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
# Leer datos
df = pd.read_csv('epa-sea-level.csv')

# Crear scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='Data')

    # Línea de regresión 1: Todos los datos
res1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
x_pred1 = np.arange(df['Year'].min(), 2051)
y_pred1 = res1.slope * x_pred1 + res1.intercept
plt.plot(x_pred1, y_pred1, 'r', label='Fit: all data')

    # Línea de regresión 2: Desde el año 2000
df_recent = df[df['Year'] >= 2000]
res2 = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
x_pred2 = np.arange(2000, 2051)
y_pred2 = res2.slope * x_pred2 + res2.intercept
plt.plot(x_pred2, y_pred2, 'green', label='Fit: from 2000')

# Etiquetas y título
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.legend()
plt.show()
```

```python
df = pd.read_csv('epa-sea-level.csv', delimiter=',')
df.head()
```

```python
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
slope1=linregress(df['Year'], df['CSIRO Adjusted Sea Level']).slope

intercept= linregress(df['Year'], df['CSIRO Adjusted Sea Level']).intercept
```

```python
x_pred = np.arange(df['Year'].min(), 2051)


y_pred = slope1 * x_pred + intercept

# Hacer el scatter plot original
plt.scatter(df['Year'], df['NOAA Adjusted Sea Level'])

# Dibujar la línea de regresión
plt.plot(x_pred, y_pred, color='red')

plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.show()
```

```python
x_pred = np.arange(df['Year'].max(), 2051)


y_pred = slope1 * x_pred + intercept

# Hacer el scatter plot original
plt.scatter(df['Year'], df['NOAA Adjusted Sea Level'])

# Dibujar la línea de regresión
plt.plot(x_pred, y_pred, color='red')

plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.show()
```
